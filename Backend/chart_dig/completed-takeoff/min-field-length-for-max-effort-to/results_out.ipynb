```python
import os
import re
os.chdir("../..")
print(os.getcwd())
from py_utils import *
```

```python
TOP_FOLDER = "completed-takeoff/min-field-length-for-max-effort-to"
```

```python
DIG_FILE_NAME = "amax-eff-uncorr-field-length.dig"
INPUT = "tof"
OUTPUT = "uncor_1"
uncorr = {}
scales = []
chart = ParseDig(f'./{TOP_FOLDER}/dig/{DIG_FILE_NAME}')
for c in  chart.curveNames():
    yVector = [row [0] for row in chart.curve(c)]
    xVector = [row [1] for row in chart.curve(c)]
    c_clean = c.replace("-", "_")
    scales.append(float(re.sub('[^0-9]','', c_clean)))
    print(f"lin(std::vector<double>{{" + ", ".join(['{0:.3f}'.format(s) for s in yVector]) + f"}}, // {c_clean.lower()}")
    print(f"std::vector<double>{{" + ", ".join(['{0:.3f}'.format(s) for s in xVector]) + f"}}, {INPUT}),")
print("std::array scale{" + ", ".join(['{0:.1f}'.format(s) for s in (scales)]) + "};")
```
