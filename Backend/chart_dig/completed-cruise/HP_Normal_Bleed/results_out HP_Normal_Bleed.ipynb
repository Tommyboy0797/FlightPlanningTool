```python
DIG_FILE_NAME = "HP_Normal_Alt_Speed.dig"
X_AXIS_NAME = "KIAS"
Y_NAME = "alt"
uncorr = {}
chart = ParseDig(f'./{TOP_FOLDER}/dig/{DIG_FILE_NAME}')
for c in chart.curveNames():
    xVector = [row [0] for row in chart.curve(c)]
    yVector = [row [1] for row in chart.curve(c)]
    c_clean = c.replace("-", "_")
    x_nm = c_clean
    print(f"const std::vector<double>{x_nm}_x{{" + ", ".join(['{0:.3f}'.format(s)for s in xVector]) + "};")
    
    print(f"const std::vector<double>{c_clean}_y{{" + ", ".join(['{0:.3f}'.format(s) for s in yVector]) + "};")
    print(f"auto hp_n_ = lin({c_clean}_y, {x_nm}_x, alt_div_1000);")
    print("\n")
```
